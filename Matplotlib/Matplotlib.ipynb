## Visualization using Matplotlib
    1) Barplot --> distribution of data (bi-variant)
    2) line chart  ---> whenever you want to see trend in the data(bivariant)
    3) Area plot ---> similar like line chart
    4) histogram ---> distribution of data[univariant]
    5) boxplot---> outlier detection(used for univariant and bivariant analysis)
    6) pie chart ---> distribution of categorical variable
    7) scatterplot---> linear relationship between two continuous variables


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## How to create Line chart

```python
x  = np.arange(0,10,0.2)
y = 2*x+5
```

```python
plt.plot(x , y)
plt.show()
```

```python
plt.plot(x , y)
plt.grid(True)
plt.show()
```

## customize the line plot

```python
plt.plot(x , y , color ='r' , linestyle = 'dashed', marker ='+' )
plt.title('Demo of Line chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Subplot

```python
plt.subplot(2 , 1 , 1)
plt.plot(x,y , color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,1,2)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python

plt.figure(figsize= (15,7))

plt.subplot(2 , 2 , 1)
plt.plot(x,y , color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,2)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,3)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,4)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

## Barplot

```python
data = {'apple': 20 ,
       'mango' :30,
       'lemon':50 ,
        'oranges' :10
       
       
       }

names_fruits = list(data.keys())
values_fruits = list(data.values())
print(names_fruits)
print(values_fruits)
```

```python
plt.bar(names_fruits , values_fruits ,color =['Red' , 'Orange','Blue','yellow'] )
plt.title('Demo of barplot')
plt.xlabel('Fruits name')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.barh(names_fruits , values_fruits ,color =['Red' , 'Orange','Blue','yellow'] )
plt.title('Demo of barplot')
plt.xlabel('count')
plt.ylabel('Fruits name')
plt.grid(True)
plt.show()
```

## Scatter plot

```python
a = [10,20,20,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
c = [1,2,3,4,3,2,1,2]

plt.scatter( a, b)
plt.show()
```

```python
plt.scatter(a, b)
plt.scatter(a , c ,c='g')
plt.legend(['a&b','a&c'])
plt.title('Demo of scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.scatter(a , b , c= 'g' , s= 200 , marker ='o')
plt.scatter (a, c , c= 'r' , s =100 , marker ='4')
plt.legend(['b' ,'c'])
plt.grid(True)
plt.title('Demo of scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.savefig(r'scatter.jpeg')
```

## How to create Histogram

```python
numbers = [10,12,16,19,20,78,67,100]
plt.hist(numbers)
plt.show()
```

```python
plt.hist(numbers , bins =[0,10,40,100] , color=['green'])# univaraint :continuous variable
plt.grid(True)
plt.title('Demo of Histogram')
plt.show()
```

```python
plt.hist(numbers , color = 'yellow' , orientation='horizontal')
plt.grid(True)
plt.title('Demo of Histogram')
plt.show()
```

## Area Plot


```python
x1 = range(1,15)
y1 =[1,4,6,8,4,5,3,2,4,1,5,6,8,7]
```

```python
plt.stackplot(x1 , y1)
plt.show()
```

```python
plt.stackplot(x1 , y1 , color = 'pink')
plt.title('Area plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# combine line chart and area plot

plt.stackplot(x1 , y1 , color ="pink")
plt.plot(x1, y1 , color = "green")
plt.legend(['line plot' ,'area chart'])
```

## Pie chart


```python
animal_name = ['Dog' ,'Cat' ,'wolf' ,'Lion']
animal_count = [50 ,45 , 60 , 70]

plt.pie(animal_count , labels= animal_name)
plt.show()
```

```python
animal_name = ['Dog' ,'Cat' ,'wolf' ,'Lion']
animal_count = [50 ,45 , 60 , 70]

plt.pie(animal_count , labels= animal_name , autopct= '%1.2f%%')
plt.show()
```

```python
plt.pie(animal_count , labels= animal_name,
       autopct= '%1.1f%%', colors = ['yellow' ,'green','pink','red'])

plt.show()
```

```python
plt.pie(animal_count , labels= animal_name,
       autopct= '%1.1f%%', colors = ['yellow' ,'green','pink','red'],
       
       explode = [0,0.1,0,0.6
                 ])

plt.show()
```

## Boxplot

```python
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discounts =[30,10,20,5,10,20,50,60,20,45]


data = list([total , order , discounts])
```

```python
plt.boxplot(total)

plt.show()
```

```python
## multiples graph in single cell using subplot

plt.figure(figsize =(10,5))
plt.subplot(1 , 3  ,1)
plt.boxplot(total)
plt.title('Total')

plt.subplot(1 , 3  ,2)
plt.boxplot(total)
plt.title('order')

plt.subplot(1 , 3  ,3)
plt.boxplot(total)
plt.title('discount')

plt.show()
```

```python
plt.boxplot(data , showmeans= True)
plt.show()
```

```python
#customization :
plt.boxplot(data , showmeans= True , vert= False)
plt.show()
```

```python
# violen Plot
plt.violinplot(data , showmeans= True)
plt.show()
```

## Example

```python
data = pd.read_csv(r'company_sales_data.csv')
data.head()
```

## Insights:
        1) max prfit in the month of Nov
        2) min profit in the month of feb
    

```python
plt.plot(data['month_number'] , data['total_profit'] , color = 'green')
plt.title('Profit distribution over months')
plt.xlabel('months'
          )
plt.ylabel('TotalProfit')
plt.grid(True)
plt.show()
```

```python
# multi line charts in the same graph

plt.plot(data['month_number'] , data['facecream'] , color = 'green')
plt.plot(data['month_number'] , data['facewash'] , color = 'orange')
plt.legend(['facecream' , 'facewash'])
plt.title('Profit distribution over months ')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize= (5,9))
plt.subplot(2,1 ,1 )
plt.plot(data['month_number'] , data['facecream'] , color = 'green')
plt.xlabel('month')
plt.ylabel('count for Facecream')

plt.subplot(2,1 ,2 )
plt.plot(data['month_number'] , data['facewash'] , color = 'orange')

plt.title('Profit distribution over months ')
plt.grid(True)
plt.show()
```

```python
# Is there any realtionship between total_units and total_profit
    # Insights : more the number of units get sold more will be the profit

plt.scatter(data['total_units'] , data['total_profit'] , color ='orange')
plt.title('Is there any realtionship between unit sold and profit?')
plt.grid(True)
plt.show()
```

```python
# histogram on toothpaste

plt.hist(data['toothpaste'] , color= 'green' , orientation='horizontal')
plt.title('Histogram')
plt.grid(True)
plt.show()
```
